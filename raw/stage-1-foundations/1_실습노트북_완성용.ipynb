## 1.5 문자열·리스트 접근하기 — 인덱싱 & 슬라이싱

- **인덱스:** 원소의 위치 번호, **0부터 시작**, 음수 인덱스는 뒤에서부터
- **인덱싱:** `자료[인덱스]` → 특정 위치의 원소 하나 반환
- **슬라이싱:** `자료[시작:끝]` → 시작 이상, 끝 **미만** 범위 반환
- 끝을 생략하면 마지막까지, 시작을 생략하면 처음부터

```
"Ready"  →  R  e  a  d  y
인덱스   →  0  1  2  3  4  (양수)
           -5 -4 -3 -2 -1  (음수)
```


````python
word    = "Ready"
numbers = [2, 4, 6, 8, 10]

# 인덱싱
print(word[0])      # R   — 첫 번째
print(word[1])      # e   — 두 번째
print(word[-1])     # y   — 뒤에서 첫 번째
print(word[-2])     # d   — 뒤에서 두 번째

print(numbers[0])   # 2
print(numbers[2])   # 6
print(numbers[-1])  # 10

# 슬라이싱
beta = [2, 4, 6, 8, 10, 12, 14]
print(beta[2:5])    # [6, 8, 10]      인덱스 2 이상 5 미만
print(beta[:3])     # [2, 4, 6]       처음~3 미만
print(beta[4:])     # [10, 12, 14]    4 이상~끝
print(beta[:])      # 전체 복사

sentence = "Hi! elice!"
print(sentence[0:3])   # Hi!
print(sentence[4:9])   # elice
````

# 2. 조건문

## 핵심 개념
- `input()` 은 사용자 입력을 받으며, 반환값은 항상 **문자열** → 숫자로 쓰려면 `int()` / `float()` 변환 필수
- **논리형(bool):** `True` / `False` 두 가지 값만 존재
- **비교 연산자:** `==` `!=` `<` `>` `<=` `>=` → 결과는 `True` or `False`
- **논리 연산자:** `and`(모두 True), `or`(하나라도 True), `not`(반전)
- **if 문:** 조건이 `True` 일 때만 들여쓰기 블록 실행
- **if-else:** True → if 블록, False → else 블록
- **if-elif-else:** 여러 조건을 순서대로 검사, 처음 True인 블록만 실행


## 2.1 입력 — `input()`
- `변수 = input("안내문구")` 형태로 사용
- 반환값이 항상 `str`이므로 산술 연산 전 형 변환 필수


````python
# input() 반환값은 str
var1 = "100"
var2 = "300"

print("문자열 더하기:", var1 + var2)            # "100300" (이어붙임!)
print("숫자로 더하기:", int(var1) + int(var2))  # 400 (덧셈)
````

### 형 변환 (Type Casting)
- `int()` `float()` `str()` `list()` 로 자료형을 원하는 형태로 바꿉니다.


````python
a = "345"
b = int("345")
print("문자열 345:", a, type(a))    # <class 'str'>
print("정수  345:", b, type(b))    # <class 'int'>

c = float("3.14")
print("실수로 변환:", c, type(c))  # <class 'float'>

d = str(1000)
print("문자열로 변환:", d, type(d)) # <class 'str'>

e = list("1008")
print("리스트로 변환:", e)          # ['1', '0', '0', '8']
````

## 2.2 논리형 & 비교·논리 연산자

| 연산자 | 의미 | 예시 | 결과 |
|--------|------|------|------|
| `==` | 같다 | `3 == 3` | `True` |
| `!=` | 다르다 | `3 != 5` | `True` |
| `<` `>` | 크기 비교 | `3 < 5` | `True` |
| `<=` `>=` | 이하/이상 | `3 >= 3` | `True` |
| `and` | 모두 True | `True and False` | `False` |
| `or` | 하나라도 True | `True or False` | `True` |
| `not` | 반전 | `not True` | `False` |


````python
# 비교 연산자
print(3 < 5)    # True
print(7 == 5)   # False
print(2 >= 10)  # False
print(5 != 10)  # True

# and: 모두 True여야 True
print(3==3 and 4<=5 and 6>2)   # True
print(3==3 and 4>5)            # False

# or: 하나라도 True면 True
print(3==4 or 4<=5 or 6<2)    # True
print(3==4 or 5==6)            # False

# not: 논리값 반전
print(not 3==4)   # True   (False → True)
print(not 3==3)   # False  (True → False)
````

### 실습 2-2 : 논리 연산자 조합 연습
- 실생활 조건을 논리 연산자로 표현해 보세요.


````python
age        = 25
height     = 170
has_ticket = True

can_ride    = age >= 18 and height >= 150 and has_ticket
is_student  = False
is_senior   = False
get_discount = is_student or is_senior
is_closed   = False
is_open     = not is_closed

print("롤러코스터 탑승 가능?", can_ride)
print("할인 적용 가능?", get_discount)
print("가게 열려있나요?", is_open)
````

## 2.3 조건문 — `if` / `if-else` / `if-elif-else`
- **들여쓰기(4칸)** 로 블록을 구분 — Python의 핵심 문법
- `if` 뒤에 반드시 `:` 콜론
- `elif` 는 원하는 만큼 체인 가능
- `else` 는 맨 마지막, 조건 없이 사용


````python
# if-else 기본
i = 3
if i == 1:
    print("i는 1입니다.")
else:
    print("i는 1이 아닙니다.")   # i == 1 이 False → else 실행

# if-elif-else: 성적 판별
score = 78

if score >= 90:
    print("학점: A")
elif score >= 75:
    print("학점: B")    # 78 ≥ 75 → B 출력
elif score >= 60:
    print("학점: C")
else:
    print("학점: F")
````

### 실습 4-1 : BMI 계산기
- `weight`, `height_cm` 변수 값을 직접 바꿔 테스트하세요.


````python
weight    = 70.0      # kg — 원하는 값으로 변경
height_cm = 175.0     # cm — 원하는 값으로 변경

height_m = height_cm / 100
bmi      = weight / (height_m ** 2)
print("BMI:", round(bmi, 2))

if bmi < 18.5:
    print("저체중입니다.")
elif bmi < 23:
    print("정상 체중입니다.")
elif bmi < 25:
    print("과체중입니다.")
else:
    print("비만입니다.")
````

# 3. 리스트

## 핵심 개념
- **리스트 생성:** `[]` 또는 `list()` — 빈 리스트부터 혼합 자료형까지
- **`append(d)`:** 맨 끝에 원소 하나 추가
- **`insert(i, d)`:** 인덱스 `i` 위치에 원소 삽입, 기존 원소는 뒤로 밀림
- **`remove(d)`:** 첫 번째로 등장하는 `d` 삭제 (없으면 ValueError)
- **`sort()`:** 원본 리스트를 오름차순 정렬, `reverse=True` 로 내림차순
- **시퀀스 자료형 공통 특징:** 인덱싱 · 슬라이싱 · `in` · `len()` · `+` · `*`
- 문자열은 `append` 불가 / 리스트는 문자 기준 `split` 불가 — 종류별 차이 주의


## 3.1 리스트 핵심 메서드

### `list.append(d)` — 맨 끝에 추가
- 한 번에 원소 **하나**만 추가 가능
- 여러 개 추가 시 `append` 반복 호출 또는 `extend()` 사용


````python
a = []
b = ["a", "b", "c"]

a.append(10)
b.append("d")

print("a:", a)   # [10]
print("b:", b)   # ['a', 'b', 'c', 'd']

shopping = []
shopping.append("사과")
shopping.append("바나나")
shopping.append("우유")
print("장바구니:", shopping)
````

### `list.sort()` — 정렬
- 숫자: 오름차순 / 문자열: 사전순 (a→z)
- `sort(reverse=True)` 로 내림차순
- 원본 리스트 **직접 변경** (반환값은 `None`)


````python
e = [6, 2, 4, 1, 3, 5]
e.sort()
print("오름차순:", e)          # [1, 2, 3, 4, 5, 6]

f = ["carrot", "apple", "banana"]
f.sort()
print("사전순:", f)            # ['apple', 'banana', 'carrot']

g = [6, 2, 4, 1, 3, 5]
g.sort(reverse=True)
print("내림차순:", g)          # [6, 5, 4, 3, 2, 1]
````

## 3.2 시퀀스 자료형 공통 특징

- **순서가 있는** 자료형 — 문자열(`str`)과 리스트(`list`) 모두 해당
- 공통 기능: 인덱싱, 슬라이싱, `in`, `not in`, `len()`, `+`, `*`
- 종류별 차이: `list.append()` 가능 ↔ `str.append()` 불가


````python
a = "Hello"
b = ["e", "l", "i", "c", "e"]

# 인덱싱 & 슬라이싱
print(a[1])      # e
print(b[2:4])    # ['i', 'c']
print(a[-1])     # o
print(b[:4])     # ['e', 'l', 'i', 'c']

# in 연산자
print("e" in a)     # True
print("E" in b)     # False (대소문자 구분)
print("z" not in a) # True

# len()
print(len("Hello World"))           # 11 (공백 포함)
print(len(["e","l","i","c","e"]))   # 5

# + 이어붙이기
print("Good" + "Day")                       # GoodDay
print(["H","e"] + ["l","l","o"])            # ['H','e','l','l','o']

# * 반복
print("Go!!" * 5)                           # Go!!Go!!Go!!Go!!Go!!
print([13, 17, 19] * 3)                     # [13,17,19, ...]
````

## 시나리오 1 : 학생 성적 관리 시스템

### 목표
- 학생 이름과 점수를 리스트로 관리
- 인덱싱으로 특정 학생 조회
- 조건문으로 합격/불합격 판별
- `sort()` 로 성적 순위 확인

### 사용 개념
- 리스트 생성 · 인덱싱 · 슬라이싱
- `for` 반복문 (미리 보기)
- `if-else` 조건문
- `sort()` · `len()`


````python
# ── 데이터 초기화 ──────────────────────────────
student_names  = ["Alice", "Bob", "Charlie", "David", "Eve"]
student_grades = [85, 72, 90, 65, 88]

print("학생 이름:", student_names)
print("학생 성적:", student_grades)
print("총 학생 수:", len(student_names))

# ── 특정 학생 조회 ──────────────────────────────
target_idx = 2   # Charlie
print("\n조회 학생:", student_names[target_idx])
print("해당 성적:", student_grades[target_idx])

# ── 합격 여부 판별 ──────────────────────────────
pass_score = 70
print("\n[ 합격 기준:", pass_score, "점 ]")

for i in range(len(student_names)):
    if student_grades[i] >= pass_score:
        status = "합격"
    else:
        status = "불합격"
    print(f"  {student_names[i]:<10} {student_grades[i]}점 → {status}")

# ── 성적 정렬 ────────────────────────────────────
sorted_grades = student_grades[:]   # 원본 보존을 위한 복사
sorted_grades.sort()
print("\n정렬된 성적:", sorted_grades)
print("최저:", sorted_grades[0], "/ 최고:", sorted_grades[-1])

top3 = sorted_grades[-3:]           # 상위 3개
print("상위 3개 성적:", top3)

# ── 학생 추가 ────────────────────────────────────
student_names.append("Frank")
student_grades.append(78)
print("\n추가 후 명단:", student_names)
print("추가 후 성적:", student_grades)

# ── 특정 점수 제거 ──────────────────────────────
student_grades.remove(65)           # 65점 제거
print("65점 제거 후:", student_grades)
````

## 시나리오 2 : 편의점 재고·매출 관리

### 목표
- 상품명·가격·재고를 리스트로 관리
- 조건문으로 재고 부족 경고 출력
- 판매 시 재고 감소 처리
- 총 매출 계산

### 사용 개념
- 리스트 인덱싱 · 값 변경
- `if-elif-else` 조건문
- 변수 누적 연산 (`+=`)
- `len()` · `range()`


````python
# ── 상품 데이터 ─────────────────────────────────
products = ["콜라", "과자", "라면", "물", "커피"]
prices   = [1500, 1200, 900, 800, 2000]   # 단가 (원)
stocks   = [10, 5, 3, 20, 7]              # 재고 (개)

print("=" * 45)
print(f"{'상품명':<8} {'단가':>6} {'재고':>6} {'상태'}")
print("=" * 45)
for i in range(len(products)):
    if stocks[i] == 0:
        status = "품절"
    elif stocks[i] <= 3:
        status = "부족"
    else:
        status = "정상"
    print(f"  {products[i]:<8} {prices[i]:>5}원  {stocks[i]:>3}개  {status}")
print("=" * 45)

# ── 판매 처리 ────────────────────────────────────
sold_idx = 2       # 판매할 상품 인덱스 (라면)
sold_qty = 2       # 판매 수량

print(f"\n[판매] {products[sold_idx]} {sold_qty}개")

if stocks[sold_idx] >= sold_qty:
    stocks[sold_idx] -= sold_qty           # 재고 감소
    revenue = prices[sold_idx] * sold_qty  # 매출
    print(f"  판매 완료! 잔여 재고: {stocks[sold_idx]}개")
    print(f"  매출: {revenue:,}원")
else:
    print(f"  재고 부족! 현재 재고: {stocks[sold_idx]}개")

# ── 총 재고 가치 계산 ────────────────────────────
total_value = 0
for i in range(len(products)):
    total_value += prices[i] * stocks[i]   # 단가 × 재고
print(f"\n총 재고 가치: {total_value:,}원")

# ── 재고 부족 상품 목록 ──────────────────────────
low_stock = []
for i in range(len(products)):
    if stocks[i] <= 3:
        low_stock.append(products[i])

if low_stock:
    print("\n재고 보충 필요:", low_stock)
else:
    print("\n모든 상품 재고 정상")
````

## 시나리오 3 : 숫자 맞히기 게임 (로직 시뮬레이션)

### 목표
- 정답 번호와 입력 번호를 비교하는 로직 구현
- 여러 후보에 대해 힌트(UP / DOWN / CORRECT) 출력
- 히스토리 리스트에 기록

### 사용 개념
- `if-elif-else` 다중 조건
- 리스트 `append` 로 히스토리 관리
- `len()` · 인덱싱


````python
# ── 게임 설정 ────────────────────────────────────
answer  = 47                          # 정답 (1~100)
guesses = [30, 60, 45, 50, 47]       # 추측 순서 (실제 게임에선 input 사용)
history = []                          # 시도 기록

print("=" * 35)
print("   숫자 맞히기 게임 시작!")
print("=" * 35)

for attempt, guess in enumerate(guesses, start=1):
    if guess == answer:
        result = "CORRECT!"
    elif guess < answer:
        result = "UP  (더 큰 수)"
    else:
        result = "DOWN (더 작은 수)"

    history.append((attempt, guess, result))   # (시도 번호, 추측값, 결과) 저장
    print(f"  시도 {attempt}: {guess:>3} → {result}")

    if guess == answer:
        print(f"\n  총 {attempt}번 만에 맞혔습니다!")
        break

# ── 히스토리 출력 ────────────────────────────────
print("\n[ 전체 기록 ]")
for rec in history:
    print(f"  {rec[0]}번째: {rec[1]} → {rec[2]}")

print(f"\n총 시도 횟수: {len(history)}회")
````

## 시나리오 4 : 간단한 로그인 & 메뉴 시스템

### 목표
- 아이디/비밀번호 검증 (조건문)
- 로그인 성공 시 메뉴 목록 표시
- 메뉴 번호로 기능 선택

### 사용 개념
- `and` 논리 연산자로 복합 조건 처리
- `if-elif-else` 메뉴 분기
- 리스트로 메뉴 항목 관리
- `in` 연산자로 유효 번호 확인


````python
# ── 계정 정보 ────────────────────────────────────
CORRECT_ID = "admin"
CORRECT_PW = "1234"

# ── 로그인 시뮬레이션 (실제 게임에선 input 사용) ─
user_id = "admin"
user_pw = "1234"

print("=" * 35)
print("       로그인 시스템")
print("=" * 35)
print(f"  아이디: {user_id}")
print(f"  비밀번호: {'*' * len(user_pw)}")   # 비밀번호 마스킹

if user_id == CORRECT_ID and user_pw == CORRECT_PW:
    print("\n  로그인 성공!")
    is_logged_in = True
elif user_id == CORRECT_ID:
    print("\n  비밀번호가 틀렸습니다.")
    is_logged_in = False
elif user_pw == CORRECT_PW:
    print("\n  아이디가 틀렸습니다.")
    is_logged_in = False
else:
    print("\n  아이디와 비밀번호 모두 틀렸습니다.")
    is_logged_in = False

# ── 메뉴 표시 ────────────────────────────────────
if is_logged_in:
    menu_items = ["회원 정보 조회", "비밀번호 변경", "로그아웃"]

    print("\n[ 메뉴 ]")
    for idx, item in enumerate(menu_items, start=1):
        print(f"  {idx}. {item}")

    # 메뉴 선택 시뮬레이션
    choice = 1    # 1~3 중 선택

    valid_choices = list(range(1, len(menu_items) + 1))   # [1, 2, 3]
    if choice in valid_choices:
        selected = menu_items[choice - 1]    # 인덱스는 0부터 → choice - 1
        print(f"\n  선택: {choice}. {selected}")

        if choice == 1:
            print("  → 회원 정보를 표시합니다.")
            print(f"     아이디: {user_id}")
        elif choice == 2:
            print("  → 비밀번호 변경 화면으로 이동합니다.")
        else:
            print("  → 로그아웃 되었습니다. 안녕히 가세요! 👋")
    else:
        print(" 잘못된 메뉴 번호입니다.")
````
